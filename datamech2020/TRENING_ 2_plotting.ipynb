```python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]
ax = plt.subplot()
n = 1  # This is our first dataset (out of 2)
t = 2 # Number of datasets
d = 6 # Number of sets of bars
w = 0.8 # Width of each bar
x_values = [t*element + w*n for element
             in range(d)]
store1_x = x_values             
plt.bar(store1_x,sales1)
ax2 = plt.subplot()
n = 2  # This is our second dataset (out of 2)
t = 2 # Number of datasets
d = 6 # Number of sets of bars
w = 0.8 # Width of each bar
x_values = [t*element + w*n for element
             in range(d)]
store2_x = x_values   
plt.bar(store2_x,sales2)

ax.set_xticks([(x1+x2)/2 for x1, x2 in zip(store1_x,store2_x)] )
ax.set_xticklabels(drinks, rotation =45)
plt.show()
```

```python
""" Результаты такие: у порівнянні з конкурентом, перший заклад має попит серед споживачів капучино; конкурент має більший попит серед споживачів єспрессо. В цілому перший заклад відомий більш своїми єксклюзивними напоями, коли як конкурент отримує успіхи в классичних напоях з кави   """
```
