```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]


visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]



plt.figure(figsize=(12,8))

ax1 = plt.subplot(1,2,1)# total page visits 
x_values = np.array(range(len(months)))
xnew = np.linspace(x_values.min(), x_values.max(), 300)  #масштабирование, оптимальное количество точек для функции
spl = make_interp_spline(x_values, visits_per_month)
power_smooth = spl(xnew)
plt.plot(xnew,power_smooth, linestyle= ':',label = 'Interpolate B-spline cubic function')
plt.plot(x_values,visits_per_month, marker= 'o',label = 'Linear function' )
plt.xlabel("Month")
plt.ylabel("Guests who visited the store per month")
plt.legend()
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)
plt.title("Frequency of veisits per month")

ax2 = plt.subplot(1,2,2)
labels_limes =['key','persian','blood'] 
plt.plot(x_values,key_limes_per_month, linestyle = '--',color = 'gray')
plt.plot(x_values,persian_limes_per_month, linestyle = '--',color = 'purple')
plt.plot(x_values,blood_limes_per_month, linestyle = '--',color = 'green')
plt.legend(labels_limes,loc = 1)
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)
plt.subplots_adjust(wspace=0.5)
plt.title("Numbers of limes of different species sold each month")
plt.show()
plt.savefig("CHALLENGE plotting_sales_limes.png")
```
